```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data_NL = pd.read_csv("GSEvaluation_NL_cap.csv")
data_MC = pd.read_csv("GSEvaluation_MC_cap.csv")
```

```python
# Option 1: replace NaN with -1
data_NL = data_NL.fillna(-1)
data_NL
# (optional, do same for data_MC)
# data_MC = data_MC.fillna(-1)
```

```python
num_threads_list = data_NL['num_threads'].unique()
spacing_list = data_NL['spacing(mm)'].unique()
perf_diff_inc_op = {}
for i in range(1, len(num_threads_list)):
    num_threads_i = num_threads_list[i]
    heatmap_col_i = data_NL[data_NL['num_threads'] == num_threads_i]
    num_threads_i_minus_1 = num_threads_list[i-1]
    heatmap_col_i_minus_1 = data_NL[data_NL['num_threads'] == num_threads_i_minus_1]
    onc_diff_list = []
    omc_diff_list = []
    for spacing in spacing_list:
        heatmap_element_i = heatmap_col_i[heatmap_col_i['spacing(mm)'] == spacing]
        onc_i = heatmap_element_i['nonlinearity test'].values[0]
        omc_i = heatmap_element_i['memory test'].values[0]
        heatmap_element_i_minus_1 = heatmap_col_i_minus_1[heatmap_col_i_minus_1['spacing(mm)'] == spacing]
        onc_i_minus_1 = heatmap_element_i_minus_1['nonlinearity test'].values[0]
        omc_i_minus_1 = heatmap_element_i_minus_1['memory test'].values[0]

        if onc_i != -1 and onc_i_minus_1 != -1:
            onc_diff = (onc_i - onc_i_minus_1) / onc_i_minus_1
            # print(f"Spacing {spacing} mm, Nonlinearity Cap Diff: {onc_diff}")
            onc_diff_list.append(onc_diff)
        if omc_i != -1 and omc_i_minus_1 != -1:
            omc_diff = (omc_i - omc_i_minus_1) / omc_i_minus_1
            # print(f"Spacing {spacing} mm, Memory Cap Diff: {omc_diff}")
            omc_diff_list.append(omc_diff)

    average_onc_diff = 100*np.sum(onc_diff_list)/len(onc_diff_list)
    average_omc_diff = 100*np.sum(omc_diff_list)/len(omc_diff_list)

    perf_diff_inc_op[f'{num_threads_i_minus_1}x{num_threads_i_minus_1} to {num_threads_i}x{num_threads_i}'] = (average_onc_diff, average_omc_diff)

print("When going from ")
for key, value in perf_diff_inc_op.items():
    print(f"{key}, nonlinear capacity increases by {value[0]:.2f}% and memory capacity increases by {value[1]:.2f}%")

print("due to increased number of outputs")
```
